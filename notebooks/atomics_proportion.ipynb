```python
import os
import pandas as pd
import re
```

### Atomic proportion

```python
data_folder = "../data/proportional"
```

```python
filter_contains = "ChiSquared"
data = []

def num_fast_slow(fn):
    with open(fn, "r") as in_xml:
        xml_str = in_xml.read()
        return xml_str.count("fast"), xml_str.count("slow")

for file in os.listdir(data_folder):
    if not filter_contains in file or not file.endswith(".xml"):
        continue
        
    match = re.search("s-([0-9]+).*_p([0-9]+)-([0-9]+)", file)
    if not match:
        continue
    
    num_fast, num_slow = num_fast_slow(os.path.join(data_folder, file))
    num_fast -= 2
    data.append(match.groups() + (num_fast, num_slow))
```

```python
data = sorted(data, key=lambda x: int(x[1]))
df = pd.DataFrame(data, columns=["size", "pods_fast", "pods_slow", "atomics_fast", "atomics_slow"])
```
