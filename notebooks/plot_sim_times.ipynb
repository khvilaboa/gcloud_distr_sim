```python
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
fn = "../data/fixed-75-perc-fast/sim_times.csv"
```

```python
df = pd.read_csv(fn, delimiter=";")
df = df[["filename", "deployment_time", "sim_time"]]
df["pods_fast"] = df["filename"].apply(lambda x: re.search("_p([0-9]+)", x).group(1))
df["pods_slow"] = df["filename"].apply(lambda x: re.search("_p.*-([0-9]+).xml", x).group(1))
df["pods_fast_slow"] = df["pods_fast"] + "x" + df["pods_slow"]
df
```

```python
fig, ax = plt.subplots(figsize=(12,7))
ax.bar(df["pods_fast_slow"], df["sim_time"], color='lightblue')
ax.set_ylim([250,550])
plt.xticks(rotation=90);
```
