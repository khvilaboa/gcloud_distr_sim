```python
import clipboard
import math
import matplotlib.pyplot as plt
import pandas as pd
import re
import os
import matplotlib

font = {'family' : 'DejaVu Sans',
        'weight' : 'normal',
        'size'   : 13}

matplotlib.rc('font', **font)
```

```python
#fn = "../data/fixed-75-perc-fast/sim_times_slow.csv"
fn = "../data/paper/sim_times.csv"
ofn = os.path.splitext(os.path.basename(fn))[0] + "{suffix}.png"
```

```python
df = pd.read_csv(fn, delimiter=";")
df.head()
```

```python
df = df[["filename", "deployment_time", "sim_time"]]
df["pods_fast"] = df["filename"].apply(lambda x: re.search("_p([0-9]+)", x).group(1))
df["pods_slow"] = df["filename"].apply(lambda x: re.search("_p.*-([0-9]+).xml", x).group(1))
# df["pods_fast_slow"] = df["pods_slow"] + "x" + df["pods_fast"]
REF = 5966
df["speedup"] = REF / df["sim_time"]
df.head()
```

```python
def gen_sc_name(row):
    if row["pods_fast"] == "0":
        return row["pods_slow"]
    elif row["pods_slow"] == "0":
        return row["pods_fast"]
    else:
        return row["pods_slow"] + "x" + row["pods_fast"]
        
df["pods_fast_slow"] = df.apply(lambda row: gen_sc_name(row), axis=1)
df.head()
```

```python
fast_inc_first = False

l1_pods, l2_pods = ("pods_slow", "pods_fast") if not fast_inc_first else ("pods_fast", "pods_slow")


bar_colors = []
for _, col in df.iterrows():
    if col[l1_pods] == "0" or col[l2_pods] == "0":
        bar_colors.append("khaki")
    elif col[l2_pods] == "1":
        bar_colors.append("lightblue")
    elif col[l1_pods] == "7":
        bar_colors.append("lightgreen")
    else:
        bar_colors.append("plum")
```

```python
def plot_times(df, colors=None, ylim1=None, ylim2=None, figsize=(12,7), ref=REF, ofn=None):
    fig, axs = plt.subplots(2, figsize=figsize)
    axs[0].bar(df["pods_fast_slow"], df["sim_time"], color=bar_colors if colors else "lightblue", edgecolor="grey")
    # ax.set_ylim([250,550])
    # ax.set_ylim([400,850])
    if ylim1:
        axs[0].set_ylim(ylim1)
    axs[0].set_ylabel("Time (s)")
    # axs[0].set_xlabel("Pods distribution (slow x fast)")
    # axs[0].set_xtickslabels(rotation=90)
    #axs[0].set_title("Simulation times")
    
    axs[1].bar(df["pods_fast_slow"], ref/df["sim_time"], color=bar_colors if colors else "lightblue", edgecolor="grey")
    if ylim2:
        axs[1].set_ylim(ylim2)
    axs[1].set_ylabel("Speedup")
    axs[1].set_xlabel("Pods distribution (slow x fast)")
    #axs[1].set_title("Speedup")
    
    for i in range(len(axs)):
        for tick in axs[i].get_xticklabels():
            tick.set_rotation(65)
    
    plt.tight_layout()
    
    if ofn:
        plt.savefig(ofn, dpi=400, facecolor="white", bbox_inches='tight')
            
    
```

```python
plot_times(df, colors=bar_colors, ofn=ofn.format(suffix=""), ref=REF)  # , ylim1=[300,700], ylim2=[8,20]
```

```python
fig, ax = plt.subplots(figsize=(11, 3.5))
#ax.bar(df["pods_fast_slow"], df["sim_time"], color=bar_colors, edgecolor="grey")
# ax.set_ylim([250,550])
ax.grid(True)
ax.set_axisbelow(True)
ax.bar(df["pods_fast_slow"], REF/df["sim_time"], color=bar_colors, edgecolor="grey")
ax.set_ylabel("Speed-up")
ax.set_xlabel("Pods distribution")
plt.tight_layout()

for tick in ax.get_xticklabels():
    tick.set_rotation(65)
    
plt.savefig(ofn.format(suffix="_speedup"), dpi=400, facecolor="white", bbox_inches='tight')
```
